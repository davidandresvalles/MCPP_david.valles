```python
#importar modulo de expresiones regular "re"

import re
import string
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
# Grab punctuation characters

string.punctuation
```

```python
# Add additional relevant found characters to strip punctuation

punctuation = string.punctuation + '–¡¿”“•\r´'
punctuation
```

```python
# Strip punctuation from string
def no_punct(string):
    transtable = string.maketrans('', '', punctuation)
    return string.translate(transtable)
```

```python
# como se ve en summary de data el type de la variable "message" es object,
# pero no quiere decir que todos sus elementos son string (contiene float) por lo es necesario convertirlos a string
data['message'] = data['message'].astype(str)
```

```python
# Add column with no punctuation

data['no_punct'] = data['message'].apply(no_punct)
```

```python
# Add column with speech column tokenized
data['tokens'] = data['no_punct'].apply(word_tokenize)
```

```python
# obtener  stopwords en español (pronombres, adverbios, articulo, preposición y conjunción) para eliminarlas de los posts.
stopwords_esp = stopwords.words('spanish')
```

```python
stopwords_esp
```

```python
def no_stopwords(tokens):
    return [w for w in tokens if w not in stopwords_esp]
```

```python
#Agragar columna sin stopwords
data['no_stopwords'] = data['tokens'].apply(no_stopwords)
```

```python
# Add column with nltk.Text object

data['nltk_Text'] = data['no_stopwords'].apply(nltk.Text)
```
