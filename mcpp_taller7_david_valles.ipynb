```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [18.0, 10.0]
plt.style.use('ggplot')
```

```python
crimes = pd.read_csv('Crimes_-_2001_to_present.csv')
```

```python
crimes
```

### Parte obligatoria

### 1.
Calcule el número de crímenes en cada Community Area en 2015. Haga un gráfico de barras que lo ilustre.

```python
crimes_by_community = crimes.groupby('Community Area')
crimes_by_community
```

```python
crimes_by_community.groups
```

```python
community_crime_count = crimes_by_community['﻿ID'].agg('count')
community_crime_count.to_frame()
```

```python
plt.title("Crímenes en cada Community Area en 2015")
plt.ylabel("Número de crimenes")
community_crime_count.plot(kind='bar');
```

### 2.
Ordene las Community Areas de acuerdo con el número de crímenes. ¿Qué Community Area (por nombre, idealmente) presenta el mayor número de crímenes? ¿El menor?

```python
community_crime_count.sort()
plt.title("Crímenes en cada Community Area en 2015")
plt.ylabel("Número de crimenes")
community_crime_count.plot(kind='bar');
```

El Community Area que presentó mas crimenes fue el 25 (Austin) con 17311 crimenes, mientras que el que presentó menos crimenes fue el 9 (Edison Park) con 257. Existen dos datos que corresponden al Community Area "0", pero este no existe.
